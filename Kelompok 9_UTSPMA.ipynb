a.1)Regresi Linier

```python
#Regresi Linier
#Import library metode machine learning.


#Siapkan data yang akan ditraining fitur (semua fitur) dan output 

#split data training dan testing

#Gunakan data trainig untuk fitting dan data testing untuk ukuran kinerja.
#Training dengan menggunakan metode Regresi Linier. 
#Lakukan berulang kali process fitting dengan mengganti hyperparameter dan mengukur kinerjanya sehingga mendaparkan model terbaik regresi linier.
#ukur kinerja setiap model pada setiap trainig dengan MSE,RMSE, R2.


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

dataset = pd.read_csv('Dataset_9.csv')

dataset.fillna(dataset.mean(), inplace=True)

X = dataset[['Building Area (sqm)', 'Number of Floors', 'Age of Building (years)', 'Average Occupants']]
y = dataset['Energy Consumption (kWh)']

random_state = 13  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=random_state)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Evaluasi model dengan MSE, RMSE, dan R2
lin_mse = mean_squared_error(y_test, y_pred)
lin_rmse = sqrt(lin_mse)
lin_r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {lin_mse}')
print(f'Root Mean Squared Error (RMSE): {lin_rmse}')
print(f'R-squared (R2): {lin_r2}')
```

a.2) Regresi Polinomial

```python
#Regresi Polinomial
#Import library metode machine learning.


#Siapkan data yang akan ditraining fitur (semua fitur) dan output 

#split data training dan testing

#Gunakan data trainig untuk fitting dan data testing untuk ukuran kinerja.
#Training dengan menggunakan metode Regresi Polinomial. 
#Lakukan berulang kali process fitting dengan mengganti hyperparameter dan mengukur kinerjanya sehingga mendaparkan model terbaik regresi linier.
#ukur kinerja setiap model pada setiap trainig dengan MSE,RMSE, R2.

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

dataset = pd.read_csv('Dataset_9.csv')

dataset.fillna(dataset.mean(), inplace=True)

X = dataset[['Building Area (sqm)', 'Number of Floors', 'Age of Building (years)', 'Average Occupants']]
y = dataset['Energy Consumption (kWh)']

random_state = 13  # Sesuai ketentuan soal
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=random_state)

degree = 2 
poly = PolynomialFeatures(degree=degree)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

model = LinearRegression()

model.fit(X_train_poly, y_train)

y_pred = model.predict(X_test_poly)

# Evaluasi model dengan MSE, RMSE, dan R2
poly_mse = mean_squared_error(y_test, y_pred)
poly_rmse = sqrt(poly_mse)
poly_r2 = r2_score(y_test, y_pred)

print(f'Degree Polynomial: {degree}')
print(f'Mean Squared Error (MSE): {poly_mse}')
print(f'Root Mean Squared Error (RMSE): {poly_rmse}')
print(f'R-squared (R2): {poly_r2}')

```

a.3) Decision Tree Regression

```python
#Decision Tree Regression
#Import library metode machine learning.


#Siapkan data yang akan ditraining fitur (semua fitur) dan output  

#split data training dan testing

#Gunakan data trainig untuk fitting dan data testing untuk ukuran kinerja.
#Training dengan menggunakan metode Decision Tree Regresi. 
#Lakukan berulang kali process fitting dengan mengganti hyperparameter dan mengukur kinerjanya sehingga mendaparkan model terbaik regresi linier.
#ukur kinerja setiap model pada setiap trainig dengan MSE,RMSE, R2.

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

dataset = pd.read_csv('Dataset_9.csv')

dataset.fillna(dataset.mean(), inplace=True)

X = dataset[['Building Area (sqm)', 'Number of Floors', 'Age of Building (years)', 'Average Occupants']]
y = dataset['Energy Consumption (kWh)']

random_state = 13  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=random_state)

model = DecisionTreeRegressor(random_state=random_state)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Evaluasi model dengan MSE, RMSE, dan R2
dt_mse = mean_squared_error(y_test, y_pred)
dt_rmse = sqrt(dt_mse)
dt_r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {dt_mse}')
print(f'Root Mean Squared Error (RMSE): {dt_rmse}')
print(f'R-squared (R2): {dt_r2}')

# Hyperparameter Tuning untuk menguji dengan kedalaman pohon yang berbeda
for max_depth in [3, 5, 10, 15, 20]:  # kedalaman pohon yang dicoba
    model = DecisionTreeRegressor(random_state=random_state, max_depth=max_depth)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    dt_mse = mean_squared_error(y_test, y_pred)
    dt_rmse = sqrt(dt_mse)
    dt_r2 = r2_score(y_test, y_pred)
    
    print(f'\nModel dengan max_depth={max_depth}:')
    print(f'Mean Squared Error (MSE): {dt_mse}')
    print(f'Root Mean Squared Error (RMSE): {dt_rmse}')
    print(f'R-squared (R2): {dt_r2}')
```

a.4) Random Forest Regression

```python
#Random Forest Regression
#Import library metode machine learning.


#Siapkan data yang akan ditraining fitur (semua fitur) dan output  

#split data training dan testing

#Gunakan data trainig untuk fitting dan data testing untuk ukuran kinerja.
#Training dengan menggunakan metode Random Forest Regression. 
#Lakukan berulang kali process fitting dengan mengganti hyperparameter dan mengukur kinerjanya sehingga mendaparkan model terbaik regresi linier.
#ukur kinerja setiap model pada setiap trainig dengan MSE,RMSE, R2.


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

dataset = pd.read_csv('Dataset_9.csv')

dataset.fillna(dataset.mean(), inplace=True)

X = dataset[['Building Area (sqm)', 'Number of Floors', 'Age of Building (years)', 'Average Occupants']]
y = dataset['Energy Consumption (kWh)']

random_state = 13  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=random_state)

model = RandomForestRegressor(random_state=random_state)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Evaluasi model dengan MSE, RMSE, dan R2
rf_mse = mean_squared_error(y_test, y_pred)
rf_rmse = sqrt(rf_mse)
rf_r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {rf_mse}')
print(f'Root Mean Squared Error (RMSE): {rf_rmse}')
print(f'R-squared (R2): {rf_r2}')

# Hyperparameter Tuning untuk menguji dengan berbagai jumlah estimator (n_estimators) dan kedalaman pohon (max_depth)
for n_estimators in [10, 50, 100, 200]:  # jumlah estimator yang dicoba
    for max_depth in [3, 5, 10, 15, 20]:  # kedalaman pohon yang dicoba
        model = RandomForestRegressor(random_state=random_state, n_estimators=n_estimators, max_depth=max_depth)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        rf_mse = mean_squared_error(y_test, y_pred)
        rf_rmse = sqrt(rf_mse)
        rf_r2 = r2_score(y_test, y_pred)
        
        print(f'\nModel dengan n_estimators={n_estimators}, max_depth={max_depth}:')
        print(f'Mean Squared Error (MSE): {rf_mse}')
        print(f'Root Mean Squared Error (RMSE): {rf_rmse}')
        print(f'R-squared (R2): {rf_r2}')
```

```python
#Random Forest Regression
#Import library metode machine learning.


#Siapkan data yang akan ditraining fitur (semua fitur) dan output  

#split data training dan testing

#Gunakan data trainig untuk fitting dan data testing untuk ukuran kinerja.
#Training dengan menggunakan metode random forest regression
#Keluarkan nilai fitur terpenting dan buat visualisasi hasil dengan chat bar.


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Dataset_9.csv')
X = df.drop(columns=['Energy Consumption (kWh)']) 
y = df['Energy Consumption (kWh)'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=13)

rf_model = RandomForestRegressor(random_state=13)
rf_model.fit(X_train, y_train)

feature_importances = rf_model.feature_importances_

# DataFrame untuk hasil fitur penting
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print("Skor Feature Importance:")
for feature, score in zip(feature_importance_df['Feature'], feature_importance_df['Importance']):
    print(f"{feature}: {score:.4f}")

# Bar Chart Fitur Penting
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis')
plt.title('Feature Importance untuk Random Forest Regression')
plt.xlabel('Skor Importance')
plt.ylabel('Feature')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Dataset_9.csv')

X = df[['Building Area (sqm)', 'Average Occupants']]
y = df['Energy Consumption (kWh)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=13)

model_results = {}

# model yang akan diuji
models = {
    'Linear Regression': LinearRegression(),
    'Polynomial Regression (degree=2)': PolynomialFeatures(degree=2),
    'Decision Tree Regression': DecisionTreeRegressor(random_state=13),
    'Random Forest Regression': RandomForestRegressor(random_state=13)
}

# training model dan evaluasi
for model_name, model in models.items():
    if model_name == 'Polynomial Regression (degree=2)':
        poly = model
        X_train_poly = poly.fit_transform(X_train)
        X_test_poly = poly.transform(X_test)
        
        #  Linear Regression untuk fitting dengan fitur polinomial
        model_lr = LinearRegression()
        model_lr.fit(X_train_poly, y_train)
        y_pred = model_lr.predict(X_test_poly)
        
        coefficients = model_lr.coef_
        feature_names = poly.get_feature_names_out(input_features=X.columns)
    
    elif model_name == 'Linear Regression':
        # Linear Regression dengan data asli
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # Koefisien fitur pada model Linear Regression
        coefficients = model.coef_
        feature_names = X.columns
    
    else:
        # fitting model lainnya langsung dengan data asli
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        if model_name in ['Decision Tree Regression', 'Random Forest Regression']:
            # ambil fitur untuk Decision Tree dan Random Forest
            feature_importance = model.feature_importances_
            feature_names = X.columns
            coefficients = feature_importance
    
    # Hitung MSE, RMSE, dan R2
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    model_results[model_name] = {'MSE': mse, 'RMSE': rmse, 'R2': r2, 'Coefficients': coefficients, 'Feature Names': feature_names}

# hasil evaluasi model
for model_name, metrics in model_results.items():
    print(f'{model_name}:')
    print(f'  MSE: {metrics["MSE"]:.4f}')
    print(f'  RMSE: {metrics["RMSE"]:.4f}')
    print(f'  R2: {metrics["R2"]:.4f}')
    print('-' * 40)

plt.figure(figsize=(8, 6))

# nentuin koefisien atau fitur penting dari model terbaik
for model_name, metrics in model_results.items():
    coefficients = metrics['Coefficients']
    feature_names = metrics['Feature Names']
    
    plt.barh(feature_names, coefficients, label=model_name)

plt.xlabel('Importance / Koefisien')
plt.ylabel('Features')
plt.title('Pebandingan Feature Importance dari Setiap Model')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('Dataset_9_2.csv')

X = data[['Temperature (Â°C)', 'Humidity (%)', 'Wind Speed (km/h)', 'Pressure (hPa)']]
y = data['Weather Condition']

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=13)

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    return accuracy, conf_matrix, precision, recall, f1

solvers = ['lbfgs', 'liblinear', 'saga']
c_values = [0.1, 1, 10]
best_logistic_model = None
best_logistic_score = 0
best_logistic_metrics = {}

for solver in solvers:
    for c in c_values:
        model = LogisticRegression(solver=solver, C=c, random_state=13, max_iter=2000)
        model.fit(X_train, y_train)
        
        accuracy, conf_matrix, precision, recall, f1 = evaluate_model(model, X_test, y_test)
        
        if accuracy > best_logistic_score:
            best_logistic_score = accuracy
            best_logistic_model = model
            best_logistic_metrics = {
                'accuracy': accuracy,
                'confusion_matrix': conf_matrix,
                'precision': precision,
                'recall': recall,
                'f1_score': f1
            }

tree_model = DecisionTreeClassifier(random_state=13)
tree_model.fit(X_train, y_train)

logistic_results = best_logistic_metrics
tree_results = evaluate_model(tree_model, X_test, y_test)

print("Logistic Regression Best Model:")
print("  Accuracy:", logistic_results['accuracy'])
print("  Confusion Matrix:\n", logistic_results['confusion_matrix'])
print("  Precision:", logistic_results['precision'])
print("  Recall:", logistic_results['recall'])
print("  F1 Score:", logistic_results['f1_score'])

print("\nHasil Decision Tree:")
print("  Accuracy:", tree_results[0])
print("  Confusion Matrix:\n", tree_results[1])
print("  Precision:", tree_results[2])
print("  Recall:", tree_results[3])
print("  F1 Score:", tree_results[4])

if tree_results[4] > logistic_results['f1_score']:
    best_model = "Decision Tree"
else:
    best_model = "Logistic Regression"

print(f"\nThe best model based on F1 Score is: {best_model}")
```
